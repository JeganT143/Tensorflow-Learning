# Imports
* Use TensorFlow to build neural networks as a sequence of layers using the <b>keras</b> framework
* keras is contained within TensorFlow
* Access keras using <b>tf.keras</b>

```python
import tensorflow as tf
import numpy as np
```

# Data

```python
x = np.array([i for i in range(-1,10)])
y = np.array([(5*i - 5) for i in range(-1,10)])
```

```python
print(f"x : {x} ")
print(f"y : {y} ")
```

# Build model

```python
# one layer with one neuron 
model = tf.keras.Sequential(
    [
        #define the input shape
        tf.keras.Input(shape=(1,)),
        
        #Add a dense layer
        tf.keras.layers.Dense(units=1)
    ]
)
```

```python
# compile the model
model.compile(optimizer='sgd' , loss= 'mean_squared_error')
```

# Training the model

```python
model.fit(x,y,epochs=500)
```

# Prediction

```python
model.predict(np.array([10]))
```
